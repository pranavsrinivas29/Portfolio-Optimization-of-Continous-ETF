```python
import yfinance as yf
import pandas as pd

# Set the parameters
ticker = 'SPY'
start_date = '2024-01-01'  
end_date = '2025-01-01'  
initial_investment = 100000  # Initial investment amount
transaction_factor = 0.9975  # Transaction cost

# Step 1: Download historical price data for SPY over 12 months
spy_data = yf.download(ticker, start=start_date, end=end_date, auto_adjust = False)
spy_data.columns = spy_data.columns.get_level_values(0)
spy_data = spy_data.drop(columns={'High', 'Low', 'Open', 'Volume'})

# Step 2: Extract the closing price on the first and last day of the period
start_price = spy_data['Close'].iloc[0]  # First day closing price
end_price = spy_data['Close'].iloc[-1]   # Last day closing price

# Step 3: Calculate how many shares you can buy on the first day with the transaction factor applied
shares_purchased = (initial_investment * transaction_factor) / start_price

# Step 4: Calculate the value of the investment at the end of the 12 months
final_value = shares_purchased * end_price  # Apply the factor on final value too


# Display the results
print(f"Start Price: {start_price:.2f}")
print(f"End Price: {end_price:.2f}")
print(f"Shares Purchased: {shares_purchased:.2f}")
print(f"Initial Value of Investment: {initial_investment:.2f}")
print(f"Final Value of Investment: {final_value:.2f}")
```
