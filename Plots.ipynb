```python
avg_metrics={
            'OLS':{
    'avg_mae':round(ols['mae'].mean(),5),
    'avg_mase':round(ols['mase'].mean(),5),
    'avg_time_consumed':round(ols['time'].mean(),5)
            },
             'LASSO':{
    'avg_mae':round(lasso['mae'].mean(),5),
    'avg_mase':round(lasso['mase'].mean(),5),
    'avg_time_consumed':round(lasso['time'].mean(),5)            
             },
            'Boosted-Trees':{
    'avg_mae':round(bt['mae'].mean(),5),
    'avg_mase':round(bt['mase'].mean(),5),
    #'avg_time_consumed':round(bt['time'].mean(),5) 
             
            },
            'Random-Forest':{
    'avg_mae':round(rf['mae'].mean(),5),
    'avg_mase':round(rf['mase'].mean(),5),
    #'avg_time_consumed':round(rf['time'].mean(),5) 
            },
            'LSTM':{
    'avg_mae':round(lstm['mae'].mean(),5),
    'avg_mase':round(lstm['mase'].mean(),5),
    'avg_time_consumed':round(lstm['time'].mean(),5) 
            }
    
}
avg_metrics
```

```python
df = pd.DataFrame(avg_metrics).T
df
```

```python
df['avg_mae'].plot(
    kind='bar', 
    color='green', 
    title='Average Time by Model',
    rot=0  # keep x-axis labels (OLS, LASSO) horizontal
)

plt.ylabel("Avg Time")
plt.tight_layout()
plt.show()
```

```python
df['avg_mase'].plot(
    kind='bar', 
    color='orange', 
    title='Average MASE by Model',
    rot=0  # keep x-axis labels (OLS, LASSO) horizontal
)

plt.ylabel("Avg Time")
plt.tight_layout()
plt.show()
```

```python
df['avg_time_consumed'].plot(
    kind='bar', 
    color='skyblue', 
    title='Average Time by Model',
    rot=0  # keep x-axis labels (OLS, LASSO) horizontal
)

plt.ylabel("Avg Time")
plt.tight_layout()
plt.show()
```
