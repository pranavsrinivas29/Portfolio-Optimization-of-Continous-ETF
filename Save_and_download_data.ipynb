```python
import yfinance as yf
import os
import pandas as pd
```

```python

def download_and_save_ticker(ticker, start="2000-01-01", end="2023-12-31", save_dir="Data/2024"):
    os.makedirs(save_dir, exist_ok=True)
    file_path = os.path.join(save_dir, f"{ticker}_training.csv")

    if os.path.exists(file_path):
        print(f"{ticker} already downloaded.")
        return

    print(f"Downloading {ticker}...")
    data = yf.download(ticker, start=start, end=end, auto_adjust=True)
    data.columns = data.columns.get_level_values(0)
    data = data.drop(columns={'High', 'Low', 'Open', 'Volume'})
    data.to_csv(file_path)
    print(f"Saved to {file_path}")
```

```python
tickers = ['SMH', 'SOXX', 'PSI', 'XSD', 'IYW', 'XLK', 'VGT', 'QQQ', 'IGM', 'IXN']
for ticker in tickers:
    download_and_save_ticker(ticker)
```
